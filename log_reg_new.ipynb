```python
# country_box_score_data.to_csv("country_box_scores.csv")

import pandas as pd
import numpy as np
country_box_score_data = pd.read_csv('country_box_scores_with_results_PP.csv', header = 0, delimiter=',')

country_box_score_data.head()
```

```python
# do one-hot encoding so that we have it

cat_vars=['Team']
for var in cat_vars:
    cat_list='var'+'_'+var
    cat_list = pd.get_dummies(country_box_score_data[var])
    data1=country_box_score_data.join(cat_list) # it can't join it right now, due to something. IT IS DUE TO ALL THE re-running!
    country_box_score_data=data1
    
cat_vars=['Team', 'Game_ID', 'Player_ID', 'Unnamed: 0', 'Name', 'Row_ID', 'Number', "Min"]
data_vars=country_box_score_data.columns.values.tolist()
to_keep=[i for i in data_vars if i not in cat_vars]
data_final=country_box_score_data[to_keep]
data_final.columns.values
```

```python
data_final_vars=data_final.columns.values.tolist()
y=['y']
X=[i for i in data_final_vars if i not in y]

print (y)
print(X)
```

```python

from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

logreg = LogisticRegression()
logreg.fit(X_train, y_train)

y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))
```

```python
from sklearn import model_selection
from sklearn.model_selection import cross_val_score
kfold = model_selection.KFold(n_splits=10, random_state=0)
modelCV = LogisticRegression()
scoring = 'accuracy'
results = model_selection.cross_val_score(modelCV, X_train, y_train, cv=kfold, scoring=scoring)
print("10-fold cross validation average accuracy: %.3f" % (results.mean()))
```
